```python
for R in LoRA_R_list:
    import torch
    from transformers import AutoTokenizer, AutoModelForSequenceClassification
    from datasets import load_dataset
    from torch.utils.data import DataLoader
    from torch.optim import AdamW
    from transformers import get_scheduler
    from tqdm.auto import tqdm
    import numpy as np
    from evaluate import load as load_metric
    from peft import LoraConfig, get_peft_model
    import matplotlib.pyplot as plt

    # 하이퍼파라미터 정의
    batch_size = 8
    learning_rate = 5e-5
    num_epochs = 5

    # 학습 및 평가 지표와 손실을 저장할 리스트 초기화
    train_accuracy_list = []
    train_f1_list = []
    train_precision_list = []
    train_recall_list = []
    train_loss_list = []

    eval_accuracy_list = []
    eval_f1_list = []
    eval_precision_list = []
    eval_recall_list = []
    eval_loss_list = []

    # 모델과 토크나이저 불러오기
    #model_name = "monologg/koelectra-small-v3-discriminator"
    model_name = "monologg/kobert"
    dataset_name = "jason9693/APEACH"  # Ensure this dataset exists on the Hugging Face Hub
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

    # 디바이스 설정 (GPU가 있으면 GPU 사용, 없으면 CPU 사용)
    device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
    model.to(device)

    # 모델에서 지원하는 모듈(레이어) 이름 리스트 생성
    supported_modules = (torch.nn.Linear, torch.nn.Embedding, torch.nn.Conv2d, torch.nn.Conv1d)

    module_names = [name for name, module in model.named_modules() if isinstance(module, supported_modules)]

    # LoRA 설정 구성
    config = LoraConfig(
        r=R,  # LoRA rank 값 (훈련 파라미터 크기 조절)
        lora_alpha=32,  # 스케일링 파라미터
        target_modules=module_names,  # LoRA를 적용할 대상 모듈 이름들
        lora_dropout=0.1,  # 드롭아웃 확률 (오버피팅 방지)
        bias="none"  # bias 파라미터 학습 여부 설정
    )

    # 모델에 LoRA 적용
    model = get_peft_model(model, config)

    # 분류기(classifier) 파라미터는 학습 가능하도록 설정
    for param in model.classifier.parameters():
        param.requires_grad = True
    model.to(device)
    if R == 2:
    # R 값이 2일 때 학습 가능한 파라미터와 그렇지 않은 파라미터 출력
        def print_parameter_details(model):
            fine_tunable_params = []
            non_fine_tunable_params = []

            for name, param in model.named_parameters():
                if param.requires_grad:
                    fine_tunable_params.append(name)
                else:
                    non_fine_tunable_params.append(name)

            print("Fine-tunable parameters:")
            for name in fine_tunable_params:
                print(name)

            print("\nNon-fine-tunable parameters:")
            for name in non_fine_tunable_params:
                print(name)

        # Apply the function to the model
        print_parameter_details(model)

    # 데이터셋 불러오기 (오류 처리 포함)
    try:
        dataset = load_dataset(dataset_name)
        print("Dataset loaded successfully.")
    except ValueError as e:
        print(f"Error loading dataset: {e}")

    # 데이터셋이 정상적으로 로드되었을 경우 데이터로더 생성 및 학습 준비
    if 'dataset' in locals():
        train_dataloader = DataLoader(dataset["train"], shuffle=True, batch_size=batch_size)
        eval_dataloader = DataLoader(dataset["test"], batch_size=batch_size)

        # 옵티마이저와 학습률 스케줄러 설정
        optimizer = AdamW(model.parameters(), lr=learning_rate)
        num_training_steps = num_epochs * len(train_dataloader)
        lr_scheduler = get_scheduler(
            name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
        )

        # 학습 루프 시작
        progress_bar = tqdm(range(num_training_steps))
        for epoch in range(num_epochs):
            model.train()
            total_train_loss = 0
            # 학습 지표 초기화

            train_metrics = {
                "accuracy": load_metric("accuracy"),
                "f1": load_metric("f1"),
                "precision": load_metric("precision"),
                "recall": load_metric("recall")
            }

            for batch in train_dataloader:
                # 배치 데이터 토크나이징 및 디바이스 할당

                inputs = tokenizer(batch['text'], padding=True, return_tensors='pt')
                inputs['labels'] = torch.tensor(batch['class'])
                inputs = {k: v.to(device) for k,v in inputs.items()}
                # 모델 순전파 및 손실 계산

                outputs = model(**inputs)
                loss = outputs.loss
                total_train_loss += loss.item() * inputs['labels'].size(0)
                loss.backward()
                # 최적화 및 스케줄러 업데이트

                optimizer.step()
                lr_scheduler.step()
                optimizer.zero_grad()
                progress_bar.update(1)
                # 예측값 계산 및 지표에 추가

                logits = outputs.logits
                predictions = torch.argmax(logits, dim=-1)
                for metric in train_metrics.values():
                    metric.add_batch(predictions=predictions, references=inputs["labels"])
            # 에포크별 평균 학습 손실 계산 및 저장

            average_train_loss = total_train_loss / len(dataset["train"])
            train_loss_list.append(average_train_loss)
            # 에포크별 학습 지표 계산 및 저장

            train_epoch_metrics = {name: metric.compute() for name, metric in train_metrics.items()}
            train_accuracy_list.append(train_epoch_metrics["accuracy"]["accuracy"])
            train_f1_list.append(train_epoch_metrics["f1"]["f1"])
            train_precision_list.append(train_epoch_metrics["precision"]["precision"])
            train_recall_list.append(train_epoch_metrics["recall"]["recall"])

            print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {average_train_loss}, Accuracy: {train_epoch_metrics['accuracy']['accuracy']}, F1: {train_epoch_metrics['f1']['f1']}, Precision: {train_epoch_metrics['precision']['precision']}, Recall: {train_epoch_metrics['recall']['recall']}")

            # 평가 루프 시작
            model.eval()
            total_eval_loss = 0
            # 평가 지표 초기화

            eval_metrics = {
                "accuracy": load_metric("accuracy"),
                "f1": load_metric("f1"),
                "precision": load_metric("precision"),
                "recall": load_metric("recall")
            }

            for batch in eval_dataloader:
                inputs = tokenizer(batch['text'], padding=True, return_tensors='pt')
                inputs['labels'] = torch.tensor(batch['class'])
                inputs = {k: v.to(device) for k,v in inputs.items()}
                with torch.no_grad():
                    outputs = model(**inputs)
                loss = outputs.loss
                total_eval_loss += loss.item() * inputs['labels'].size(0)

                logits = outputs.logits
                predictions = torch.argmax(logits, dim=-1)
                for metric in eval_metrics.values():
                    metric.add_batch(predictions=predictions, references=inputs["labels"])
            # 에포크별 평균 평가 손실 계산 및 저장

            average_eval_loss = total_eval_loss / len(dataset["test"])
            eval_loss_list.append(average_eval_loss)
            # 에포크별 평가 지표 계산 및 저장

            eval_epoch_metrics = {name: metric.compute() for name, metric in eval_metrics.items()}
            eval_accuracy_list.append(eval_epoch_metrics["accuracy"]["accuracy"])
            eval_f1_list.append(eval_epoch_metrics["f1"]["f1"])
            eval_precision_list.append(eval_epoch_metrics["precision"]["precision"])
            eval_recall_list.append(eval_epoch_metrics["recall"]["recall"])

            print(f"Epoch {epoch+1}/{num_epochs} - Eval Loss: {average_eval_loss}, Accuracy: {eval_epoch_metrics['accuracy']['accuracy']}, F1: {eval_epoch_metrics['f1']['f1']}, Precision: {eval_epoch_metrics['precision']['precision']}, Recall: {eval_epoch_metrics['recall']['recall']}")

    else:
        print("Dataset is not loaded. Cannot proceed with training.")

    # 학습 및 평가 지표 그래프 출력
    epochs = range(1, num_epochs + 1)
    plt.figure(figsize=(12, 18))

    plt.subplot(3, 2, 1)
    plt.plot(epochs, train_accuracy_list, 'b', label='Train Accuracy')
    plt.plot(epochs, eval_accuracy_list, 'r', label='Eval Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(3, 2, 2)
    plt.plot(epochs, train_f1_list, 'b', label='Train F1 Score')
    plt.plot(epochs, eval_f1_list, 'r', label='Eval F1 Score')
    plt.title('F1 Score over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('F1 Score')
    plt.legend()

    plt.subplot(3, 2, 3)
    plt.plot(epochs, train_precision_list, 'b', label='Train Precision')
    plt.plot(epochs, eval_precision_list, 'r', label='Eval Precision')
    plt.title('Precision over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Precision')
    plt.legend()

    plt.subplot(3, 2, 4)
    plt.plot(epochs, train_recall_list, 'b', label='Train Recall')
    plt.plot(epochs, eval_recall_list, 'r', label='Eval Recall')
    plt.title('Recall over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Recall')
    plt.legend()

    plt.subplot(3, 2, 5)
    plt.plot(epochs, train_loss_list, 'b', label='Train Loss')
    plt.plot(epochs, eval_loss_list, 'r', label='Eval Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()
    print("R={} 일때 가장 높았던 시험데이터 f1 점수 값 : {}".format(R, np.max(eval_f1_list)))
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_dataset
from torch.utils.data import DataLoader
from torch.optim import AdamW
from transformers import get_scheduler
from tqdm.auto import tqdm
import numpy as np
from evaluate import load as load_metric
from peft import LoraConfig, get_peft_model
import matplotlib.pyplot as plt

# Define hyperparameters
batch_size = 8
learning_rate = 5e-5
num_epochs = 5
R = 4  # LoRA r값 고정

# Dataset
dataset_name = "jason9693/APEACH"
dataset = load_dataset(dataset_name)
tokenizer = AutoTokenizer.from_pretrained("monologg/kobert")

# Get all Linear (Dense) layer names
model_base = AutoModelForSequenceClassification.from_pretrained("monologg/kobert", num_labels=2)
dense_layer_names = [name for name, module in model_base.named_modules() if isinstance(module, torch.nn.Linear)]

# Iterate over each dense layer name
for target_layer in dense_layer_names:
    print(f"\n==== Training with LoRA applied to: {target_layer} ====")

    # Reload model
    model = AutoModelForSequenceClassification.from_pretrained("monologg/kobert", num_labels=2)
    model.to("cuda" if torch.cuda.is_available() else "cpu")

    # Apply LoRA
    config = LoraConfig(
        r=R,
        lora_alpha=32,
        target_modules=[target_layer],
        lora_dropout=0.1,
        bias="none",
        task_type="SEQ_CLS"
    )
    model = get_peft_model(model, config)
    model.to("cuda" if torch.cuda.is_available() else "cpu")

    # Prepare data
    def tokenize_fn(example):
        return tokenizer(example["text"], padding="max_length", truncation=True, max_length=128)

    tokenized = dataset.map(tokenize_fn, batched=True)
    tokenized.set_format(type="torch", columns=["input_ids", "attention_mask", "class"])
    tokenized = tokenized.rename_column("class", "labels")

    train_loader = DataLoader(tokenized["train"], shuffle=True, batch_size=batch_size)
    test_loader = DataLoader(tokenized["test"], batch_size=batch_size)

    # Optimizer and scheduler
    optimizer = AdamW(model.parameters(), lr=learning_rate)
    num_training_steps = num_epochs * len(train_loader)
    lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)

    # Lists to save metrics
    train_loss_list, eval_loss_list = [], []
    train_f1_list, eval_f1_list = [], []
    train_accuracy_list, eval_accuracy_list = [], []

    # Metric loader
    accuracy_metric = load_metric("accuracy")
    f1_metric = load_metric("f1")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        accuracy_metric.reset()
        f1_metric.reset()

        for batch in train_loader:
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = outputs.loss
            total_loss += loss.item()

            loss.backward()
            optimizer.step()
            lr_scheduler.step()
            optimizer.zero_grad()

            preds = torch.argmax(outputs.logits, dim=1)
            accuracy_metric.add_batch(predictions=preds, references=batch["labels"])
            f1_metric.add_batch(predictions=preds, references=batch["labels"])

        avg_train_loss = total_loss / len(train_loader)
        train_loss_list.append(avg_train_loss)
        train_accuracy_list.append(accuracy_metric.compute()["accuracy"])
        train_f1_list.append(f1_metric.compute(average="macro")["f1"])

        # Evaluation
        model.eval()
        total_loss = 0
        accuracy_metric.reset()
        f1_metric.reset()

        with torch.no_grad():
            for batch in test_loader:
                batch = {k: v.to(device) for k, v in batch.items()}
                outputs = model(**batch)
                loss = outputs.loss
                total_loss += loss.item()

                preds = torch.argmax(outputs.logits, dim=1)
                accuracy_metric.add_batch(predictions=preds, references=batch["labels"])
                f1_metric.add_batch(predictions=preds, references=batch["labels"])

        avg_eval_loss = total_loss / len(test_loader)
        eval_loss_list.append(avg_eval_loss)
        eval_accuracy_list.append(accuracy_metric.compute()["accuracy"])
        eval_f1_list.append(f1_metric.compute(average="macro")["f1"])

        print(f"Epoch {epoch + 1}/{num_epochs}")
        print(f"Train Loss: {avg_train_loss:.4f}, Accuracy: {train_accuracy_list[-1]:.4f}, F1: {train_f1_list[-1]:.4f}")
        print(f"Eval  Loss: {avg_eval_loss:.4f}, Accuracy: {eval_accuracy_list[-1]:.4f}, F1: {eval_f1_list[-1]:.4f}")

    # Plotting
    epochs = list(range(1, num_epochs + 1))
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_f1_list, label="Train F1")
    plt.plot(epochs, eval_f1_list, label="Eval F1")
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title(f"F1 Score for {target_layer}")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_loss_list, label="Train Loss")
    plt.plot(epochs, eval_loss_list, label="Eval Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Loss for {target_layer}")
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"[{target_layer}] 최고 F1 점수 (검증): {max(eval_f1_list):.4f}")
```
